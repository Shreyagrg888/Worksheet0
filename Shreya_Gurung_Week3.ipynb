```python
import pandas as pd

# Load the Titanic dataset
dataset = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

# Fill missing 'Age' values with the median of the 'Age' column
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].median())

# 1. Add a column 'Fare per year' containing the fare divided by the age
dataset['Fare per year'] = dataset['Fare'] / dataset['Age']

# 2. Subset rows where fare per year is higher than 5
fare_per_age = dataset[dataset['Fare per year'] > 5]

# 3. Sort high fare age by descending 'Fare per year'
high_fare_age_srt = fare_per_age.sort_values(by='Fare per year', ascending=False)

# 4. Select only the Name and fare per year columns
result = high_fare_age_srt[['Name', 'Fare per year']]

# 5. Look at the result
print(result.head())
```

```python
#Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class

dataset = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

#1
dataset['Fare per class'] = dataset['Fare']/dataset['Pclass']

#2
adult_male = dataset[(dataset['Sex']=='male')&(dataset['Age']>=18)]

#3
adult_males_srt = adult_male.sort_values(by='Fare per class', ascending = False)

#4
result = adult_males_srt[['Name','Age', 'Fare per class']]

#5
print(result.head())
```

```python
#Exploratory Data Analysis with group by method
#What percent of the total fare revenue came from each passenger class?


dataset = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
#1
total_fare = dataset['Fare'].sum()

#2
first_class_fare = dataset[dataset['Pclass']==1]['Fare'].sum()

#3
second_class_fare = dataset[dataset['Pclass']==2]['Fare'].sum()

third_class_fare = dataset[dataset['Pclass']==3]['Fare'].sum()

#4
total_fare_per_class = [first_class_fare,second_class_fare,third_class_fare]

#5
proportions_by_class = [fare / total_fare for fare in total_fare_per_class]

#output
print("Total fare by class", total_fare_per_class)
print("Proportions_by_class", proportions_by_class)




```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?

# 1. Create a new column 'AgeGroup' that categorizes passengers into "child", "adult", and "senior".
dataset['AgeGroup'] = pd.cut(dataset['Age'], bins=[-1, 18, 64, float('inf')], labels=['child', 'adult', 'senior'])

# 2. Calculate the total number of passengers on the Titanic.
total_passengers = len(dataset)

# 3. Count the number of passengers in each age group.
age_group_counts = dataset['AgeGroup'].value_counts()

# 4. Divide the count of each age group by the total number of passengers to get the proportion.
age_group_proportions = age_group_counts / total_passengers

# 5. Display the proportion as a percentage.
age_group_percentages = age_group_proportions * 100

# Display the results
age_group_percentages


```
