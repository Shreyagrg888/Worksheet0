```python
#Data imputations
#1
import pandas as pd
dataset = pd.read_csv("/content/drive/MyDrive/medical_students_dataset.csv")
dataset.head()
```

```python
#2
dataset_info = dataset.info()
print(dataset_info)

null_values = dataset.isnull()
print("Null values in each column: ")
print(null_values)
```

```python
#3

columns = dataset.select_dtypes(include=['object']).columns
#using for loop to get through each columns
for col in columns:
  #calculation of the most frequent value (mode) of the column
  mode = dataset[col].mode()[0]
  #filling the missing column with mode
  dataset[col].fillna(mode, inplace = True)
  print(f"Missing value in {col} filled with mode: {mode}")

#I used mode to fill the missing value in categorical columns because they are not numeric and they include text value
```

```python
numeric_columns = dataset.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_columns:
  median = dataset[col].median()
  dataset[col].fillna(median,inplace = True)
  print(f"Missing value in {col} filled with median: {median}")

  #I used median to fill out missing value in numeric columns because median is less sensitive to outliner compared to mean.
```

```python
#4
duplicate = dataset.duplicated().sum()
print(f"Number of duplicate rows: {duplicate}")

if duplicate > 0:
  dataset = dataset.drop_duplicates()
  print("Duplicates are removed")
else:
  print("There were no duplicates")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tdf = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

#Breaking the age column in 5 quantitles

age = pd.qcut(tdf['Age'],5, labels = ['Q1','Q2','Q3','Q4','Q5'])
tdf['Age_group'] = age

#grouping pclass and age group and mean of survival rate

mean_survival_by_age = tdf.groupby(['Pclass','Age'])['Survived'].mean().reset_index()

plt.figure(figsize =(100,100))
sns.barplot(x='Age', y='Survived', hue='Pclass', data=mean_survival_by_age)

plt.title('Survival rate by age group and passanger class')
plt.xlabel('Age')
plt.ylabel('Survival')

plt.show()


```
